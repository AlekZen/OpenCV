```python
import cv2
import os
import imutils
import tkinter as tk
from datetime import datetime
```

```python
import tkinter as tk
persona = ''

def show_entry_fields():
    show_entry_fields.persona=(e1.get())
    print(persona)
    print("Nombre: %s" % e1.get())
    

master = tk.Tk()
tk.Label(master, 
         text="Nombre").grid(row=0)

e1 = tk.Entry(master)


e1.grid(row=0, column=1)


tk.Button(master, 
          text='Ejecutar con este', command=show_entry_fields).grid(row=3, 
                                                       column=1, 
                                                       sticky=tk.W, 
                                                       pady=4)

tk.mainloop()
```

```python
persona= show_entry_fields.persona
dataPath = 'datos'
datosPersona = dataPath + '/' + persona
print(datosPersona)
```

```python
if not os.path.exists(datosPersona):
        print('Nueva carpeta para:' + datosPersona)
        os.makedirs(datosPersona)
```

```python
fecha = datetime.now()
print(fecha)
```

```python
cap = cv2.VideoCapture(0)
faceClassif = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
count=300
while True:
    ret, frame = cap.read()
    if ret==False: break
    frame = imutils.resize(frame,width=640)
    gray   = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    auxFrame = frame.copy()
    faces = faceClassif.detectMultiScale(gray,1.3,5)
    
    for(x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,+h),(0,255,0),2)
        fecha = str(datetime.now())
        rostro = auxFrame[y:y+h,x:x+w]
        rostro = cv2.resize(rostro,(150,150),interpolation=cv2.INTER_CUBIC)
        cv2.imwrite(datosPersona +'/rostro_{}.jpg'.format(count),rostro)
        count = count +1
        #print(datosPersona+'/_rostro{}.jpg'.format(count),rostro)
    cv2.imshow('Capturando',frame)

    k = cv2.waitKey(1)
    if k == 27 or count >= 600:
        break
cap.release()
cv2.destroyAllWindows()
```
